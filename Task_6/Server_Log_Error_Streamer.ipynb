- Task: 

● Server Log Error Streamer: The backend team has server log files too large to load 
into memory. Build a streaming system that reads line by line, yields only critical error 
entries, groups them by error type, and shows a live running count — without loading the 
full file at once. 

```python
# creating a server log file for inspection 
# making a generator to read the  file line by line

def read_line(file):
    with open(file , "r") as log:
        for line in log:
            yield line
```

```python
#  generator for critical ine

def critical_line(lines):
    for line in lines:
        if 'CRITICAL' in line:
            yield line
```

```python
#  for counting the errors 
def counting_error(file):
    error_count = {}

    logs = read_line(file)
    crit  = critical_line(logs)

    for error in crit:
        error_type = error.replace("CRITICAL " , "")
        error_count[error_type] = error_count.get(error_type , 0) + 1

        print("New Critical Error" , error_type)
        print("Current Count" , error_count)
        print()
```

```python
counting_error("server.log")
```
