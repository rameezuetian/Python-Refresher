- Task: 

● Order Discount Pipeline: An online store needs a data pipeline that takes a raw list of 
customer orders, applies category-based discount rules, filters out orders below a 
minimum threshold, and returns a processed summary with total revenue — using no 
traditional `for` loops anywhere. 

```python
orders = [
    {"id": 1, "category": "electronics", "price": 1000},
    {"id": 2, "category": "clothing", "price": 200},
    {"id": 3, "category": "electronics", "price": 500},
    {"id": 4, "category": "grocery", "price": 80},
    {"id": 5, "category": "clothing", "price": 150}
]
```

```python
discount_rules = {
    "electronics": 0.10,
    "clothing": 0.20,
    "grocery": 0.05
}
```

```python
discounted_orders = list(
    map(
        lambda order: {
            "id": order["id"],
            "category": order["category"],
            "original_price": order["price"],
            "discounted_price": order["price"] * (1 - discount_rules.get(order["category"]))
        },
        orders
    )
)

```

```python
# filter the small order 
filter_order = list(filter(lambda x: x['discounted_price'] >= 100, discounted_orders))
```

```python
print(filter_order)
```

```python
# Calculate the total revenue 
from functools import reduce 
total_revenue  = reduce(lambda total , order : total + order['discounted_price'] , filter_order , 0)
```

```python
summary = {
    "processed_orders": filter_order,
    "total_revenue": total_revenue
}

print(summary)
```
