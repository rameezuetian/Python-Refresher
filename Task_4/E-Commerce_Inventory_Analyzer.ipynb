```python
#  Bubble sort
arr = [64, 34, 25, 12, 22, 11, 90]

for i in range(len(arr)):
    for j in range(len(arr) - 1):
        if arr[j] > arr[j + 1]:
            arr[j]  , arr[j + 1 ] = arr[j + 1]  , arr[j]
print(arr)

```

```python
#  Selection Sort

arr = [64, 25, 12, 22, 11]

for i in range(len(arr)):
    min_index  = i 
    for j in range(i+1  , len(arr)):
        if arr[j] < arr[min_index]:
            min_index = j
    arr[i] , arr[min_index] = arr[min_index] , arr[i]
print(arr)
```

```python
#  insetrion sort
arr = [12, 11, 13, 5, 6]
for i in range(1 , len(arr)):
    key = arr[i]
    j = i - 1
    while j >= 0 and key < arr[j]:
        arr[j + 1] = arr[j]
        j -= 1
    arr[j + 1] = key 
print(arr)
```

```python
# binary search
def binary_search(arr, target):
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = left + (right - left) // 2

        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return -1

arr = [2, 4, 6, 8, 10]
target = 6
result = binary_search(arr, target)
print(result)
```

# Task: 
● E-Commerce Inventory Analyzer: A shop owner has a messy product list with 
duplicates, missing prices, and mixed categories. Clean the data, group by category, find 
the top 5 most expensive items per category, and produce a summary dictionary — 
written in a Pythonic style using comprehensions throughout. 

```python
products = [
    {"name": "Laptop", "price": 900, "category": "Electronics"},
    {"name": "Mouse", "price": 20, "category": "Electronics"},
    {"name": "Keyboard", "price": 45, "category": "Electronics"},
    {"name": "Laptop", "price": 900, "category": "Electronics"},  # duplicate
    {"name": "Desk", "price": 200, "category": "Furniture"},
    {"name": "Chair", "price": 120, "category": "Furniture"},
    {"name": "Lamp", "price": None, "category": "Furniture"},     # missing price
    {"name": "Sofa", "price": 500, "category": "Furniture"},
    {"name": "TV", "price": 800, "category": "Electronics"},
    {"name": "Phone", "price": 700, "category": "Electronics"},
    {"name": "Tablet", "price": 400, "category": "Electronics"},
]
```

```python
#  remove none and duplicates

seen = set()

clean_data = [p for p in products if p['price'] is not None and (p['name'] , p['category']) not in seen and not seen.add((p['name'] , p['category']))]
```

```python
print(clean_data)
```

```python
#  group product by category
grouped = {}
for p in clean_data:
    category = p['category']
    if category not in grouped:
        grouped[category] = []
    grouped[category].append(p)
print(grouped)
```

```python
#  find top 5 item 
top_item = sorted(clean_data , key=lambda x: x['price'], reverse=True)
print(top_item[:5])
```

```python
#  create summary
summary = {
    cat: {
        "count": len(items),
        # sort each category's items by price and grab top 5 names
        "top_items": [item["name"] for item in sorted(items, key=lambda x: x["price"], reverse=True)[:5]]
    }
    for cat, items in grouped.items()
}
```

```python
print(summary)
```

- task4.1 HR Employee Ranker: HR wants employees ranked by performance score, then by 
experience as a tiebreaker, then by name alphabetically. The system must also support 
quickly searching for a specific employee by name from the sorted list.

```python
employees = [
    {"name": "Ali", "score": 90, "experience": 5},
    {"name": "Sara", "score": 95, "experience": 3},
    {"name": "Ahmed", "score": 90, "experience": 7},
    {"name": "Zara", "score": 95, "experience": 5},
    {"name": "Bilal", "score": 88, "experience": 6}
]
```

```python
#  Sort Employees
sorted_employees = sorted(employees , key = lambda x: (-x['score'] , -x['experience'] , x['name']))
print(sorted_employees)
```

```python
# Ranking 
print("Employee Ranking:\n")

for i, emp in enumerate(sorted_employees, start=1):
    print(i, emp["name"], emp["score"], emp["experience"])
```

```python
#  for binary search create  list of names
names = [emp['name'] for emp in sorted_employees]
```

```python
def search_employee(name):

    left = 0
    right = len(names) - 1

    while left <= right:

        mid = (left + right) // 2

        if names[mid] == name:
            return sorted_employees[mid]

        elif names[mid] < name:
            left = mid + 1

        else:
            right = mid - 1

    return None
```

```python
result = search_employee("Ali")

if result:
    print("\nEmployee found:", result)
else:
    print("\nEmployee not found")
```
