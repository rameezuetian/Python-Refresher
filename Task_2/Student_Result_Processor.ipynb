- Task: 

● Student Result Processor: The academic department has CSV files of student records 
in a folder. Some files are corrupted, some are missing, and some have invalid data. 
Build a system that reads all files safely, flags failed students, and writes a clean results 
file without ever crashing.  

```python
import os
import csv

folder = "student_folder"
clean_data = []

for filename in os.listdir(folder):

    if filename.endswith(".csv"):

        path = os.path.join(folder, filename)

        try:
            with open(path, "r") as file:
                reader = csv.DictReader(file)

                for row in reader:
                    try:
                        name = row["name"].strip()
                        marks = float(row["marks"])

                        if name == "":
                            continue

                        if marks < 40:
                            status = "Fail"
                        else:
                            status = "Pass"

                        clean_data.append([name, marks, status])

                    except:
                        # Skip invalid rows
                        continue

        except:
            # Skip bad files
            print("Skipping file:", filename)


# Write clean result file
with open("clean_results.csv", "w", newline="") as out:
    writer = csv.writer(out)
    writer.writerow(["Name", "Marks", "Status"])
    writer.writerows(clean_data)

print("Done. Clean results saved.")
```
