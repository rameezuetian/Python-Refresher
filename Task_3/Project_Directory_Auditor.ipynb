Task: 

● Project Directory Auditor: The DevOps team needs a tool that scans a given project 
folder, collects each file's name, size, type, and last modified date, then generates a 
timestamped audit report. Files not modified in over 30 days must be flagged, and a 
separate penalties report must list any calculated overage fines. Package the scanning 
logic as a custom importable module that the audit script imports and uses.

```python
import os
import datetime


def scan_tool(file):

    # check file extension
    if not (file.endswith(".csv") or file.endswith(".txt")):
        print("Please enter the correct extension file")
        return

    start = datetime.datetime.now()

    print("Start Time:", start)

    print("Current working directory:", os.getcwd())

    # check file exists
    if os.path.exists(file):

        print("File exists")

        print("File stats:", os.stat(file))

    else:
        print("File does not exist")

    end = datetime.datetime.now()

    print("End Time:", end)

    total = end - start

    print("Total time taken:", total)

    if os.path.exists(file):

        modified_time = datetime.datetime.fromtimestamp(os.path.getmtime(file))

        now = datetime.datetime.now()

        difference = now - modified_time

        print("Last Modified:", modified_time)
        print("Days since modified:", difference.days)

        if difference.days > 30:
            print("FLAG: File not modified for 30 days")
        else:
            print("File recently modified")

    else:
        print("File not found")
```

```python
info = scan_tool("clean_results.csv")
```
