# OOP: Inheritance, Polymorphism & Dunder: 

Concepts: 

● Single Inheritance — extending a parent with `super()`  

● Multilevel Inheritance — parent → child → grandchild chains  

● Multiple Inheritance — two or more parent classes  

● MRO (Method Resolution Order) — how Python picks which method to run  

● Method Overriding — child replaces a parent method  

● Method Overloading — Python's approach with `*args` / `**kwargs`  

● Duck typing in practice  

● `__len__`, `__eq__`, `__lt__`, `__gt__` — size and comparison  

● `__add__`, `__sub__`, `__mul__` — operator overloading  

● `__getitem__`, `__setitem__` — subscript access  

● `__iter__`, `__next__` — making objects iterable  

● `abc` module and `@abstractmethod` — enforcing contracts across subclasses 

- Task: 

● Library Management System: A library needs to manage books, members, and 
borrowing records. Books are borrowed and returned. Members have limits. Overdue 
books trigger fines. Different membership tiers have different rules and privileges. Use 
single and multilevel inheritance for tiers, method overriding for fine logic, and dunder 
methods for display. 

```python
# make a book class
class Book:
    def __init__(self , title , author):
        self.title = title
        self.author = author
        self.is_borrowed = False

    def __str__(self):
        status = "Borrowed" if self.is_borrowed else "Available"
        return f"Title: {self.title}, Author: {self.author}, Status: {status}"
```

```python
# member class
class Member:
    def __init__(self, name):
        self.name = name
        self.borrowed_books = []
        self.limit = 2

    def borrow_book(self, book):
        if len(self.borrowed_books) > self.limit:
            print(f"{self.name} has reached the borrowing limit.")
            return

        if book.is_borrowed:
            print(f"{book.title} is already borrowed.")
            return

        book.is_borrowed = True
        self.borrowed_books.append(book)
        print(f"{self.name} borrowed {book.title}.")

    def return_book(self , book):
        if book in self.borrowed_books:
            book.is_borrowed = False
            self.borrow_book.remove(book)
            print(f"{self.name} returned {book.title}.")

    def calculate_fine(self , days):
        return days *2
    
    def __str__(self):
        return f"Member {self.name} , Borrowed Books is {len(self.borrowed_books)}"
        
```

```python
# member tiers
class SilverMember(Member):

    def __init__(self, name):
        super().__init__(name)
        self.limit = 3
```

```python
class GoldMember(SilverMember):

    def __init__(self, name):
        super().__init__(name)
        self.limit = 5
```

```python
class GoldMember(SilverMember):

    def __init__(self, name):
        super().__init__(name)
        self.limit = 5

    def calculate_fine(self, days):
        return days * 1
```

```python
class BorrowRecord:

    def __init__(self, member, book, days):
        self.member = member
        self.book = book
        self.days = days

    def show_fine(self):
        fine = self.member.calculate_fine(self.days)
        print("Fine for", self.member.name, "=", fine)

    def __str__(self):
        return f"{self.member.name} borrowed '{self.book.title}' for {self.days} days"
```

```python
book1 = Book("Python Basics", "John")
book2 = Book("AI Fundamentals", "Sara")

member1 = Member("Ali")
member2 = GoldMember("Ahmed")

member1.borrow_book(book1)
member2.borrow_book(book2)

print(book1)
print(book2)

record1 = BorrowRecord(member1, book1, 5)
record1.show_fine()

record2 = BorrowRecord(member2, book2, 5)
record2.show_fine()
```
