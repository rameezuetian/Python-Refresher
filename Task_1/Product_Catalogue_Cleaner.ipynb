## Task:
● Product Catalogue Cleaner: A supplier sends a raw product list with mixed types,
duplicates, and missing values. Clean the data, convert all prices to float, group products
into a dictionary by category, and print a neat summary.

```python
raw_products = [
    {"name": " Apple ", "price": "100", "category": "Fruits"},
    {"name": "banana", "price": 50, "category": "fruits"},
    {"name": "Mango", "price": "120.5", "category": "Fruits"},
    {"name": "apple", "price": "100", "category": "Fruits"}, 
    {"name": "Carrot", "price": None, "category": "Vegetables"},
    {"name": "Potato", "price": "30", "category": "vegetables"},
    {"name": "", "price": "60", "category": "Snacks"},        
    {"name": "Chips", "price": "abc", "category": "Snacks"},  
    {"name": "Milk", "price": 80.0, "category": None},       
    {"name": "Bread", "price": "40", "category": "Bakery"},
    {"name": "bread", "price": "40", "category": "Bakery"},   
    {"name": None, "price": "25", "category": "Snacks"}
]
```

```python
valid_products = []
seen = set()

for raw in raw_products:
    name = raw.get("name")
    category = raw.get("category")
    price = raw.get("price")

    if not name or not category:
        continue

    name = name.strip().lower()
    category = category.strip().lower()

    if not name:
        continue

    try:
        price = float(price)
    except (TypeError, ValueError):
        continue

    unique_key = (name, category)
    if unique_key in seen:
        continue

    seen.add(unique_key)

    valid_products.append({
        "name": name,
        "price": price,
        "category": category
    })


# GROUPING STEP (outside loop)
catalogue = {}

for product in valid_products:
    category = product["category"]
    catalogue.setdefault(category, []).append(product)


# PRINT SUMMARY
for category, products in catalogue.items():
    print(f"\nCategory: {category}")
    total_price = 0

    for item in products:
        print(f"  - {item['name']}: {item['price']}")
        total_price += item["price"]

    print(f"Total items: {len(products)}")
    print(f"Average price: {round(total_price / len(products), 2)}")
```
